## EDA:
1. There was presnce of null values, whihc was dropped from the dataset.
2. From a deeper look at the description of ananlysis and histogarm that we have plotted we can create two vizualizations,
   i.e. Square Footage per County and Number of floors per Usage Category.
3. Building Dataset consists of 22 columns and 8862 rows


   
   For Vizualization we intend to use altair object which is defined as alt.

```python
chart1 = alt.Chart.from_dict(
{
  'data': {"url": "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_bcubcg_fall2022/main/data/building_inventory.csv"
  }, 
  'width' : 250,
  'height' : 1000,
  'title': "Sqaure Footage per County",
  'mark': {"type":"bar", "tooltip": {"content": "data"}},
  'encoding': {
    "x": {"aggregate": "sum", "field": "Square Footage", "type": "quantitative", "title": "Square Footage"},
    "y": {"field": "County", "type": "nominal"},
    "color": {"value": "green"}
  }
}).interactive()
```

```python
chart1
```

This Plot is an Interactive plot of Square Footage per counties in the state of Illinois,From the look at ot we find Cook County has higher number of Square Footage folowed by Champaign and Sangamon.

```python
myJekyllDir = r'C:\Users\PC\kmadhura08.github.io\assets\json\json'
```

```python
chart1.properties(width='container').save(myJekyllDir+'sq_footage.json')
# note the "container" property allows us to resize the visualization 
# when the page is resized
```

```python
chart2 = alt.Chart.from_dict(
{
  'data': {"url": "https://uiuc-ischool-dataviz.github.io/spring2019online/week02/building_inventory.csv"
  }, 
  'width' : 800,
  'height' : 350,
  'title': "Number of Floors per Usage Category",
  'mark': {"type":"bar", "tooltip": {"content": "data"}},
  'encoding': {
    "x": {"aggregate": "sum", "field": "Total Floors", "type": "quantitative", "title": "Number of Floors"},
    "y": {"field": "Usage Description", "type": "nominal"},
    "color": {"value": "blue"}
  }
}).interactive()
```

```python
chart2
```

This Plot is an Interactive plot of Number of Floors per Usage Category in the state of Illinois,From the look atiot we find Storage has high number of flors that are under use which makes sense.

```python
chart2.properties(width='container').save(myJekyllDir+'floors.json')
# note the "container" property allows us to resize the visualization 
# when the page is resized
```
